# Model Traning 

Train Test Spilt

```python
from sklearn.model_selection import train_test_split
```

```python
array = sonar.values
x = array[:,0:60]
y = array[:,60]
validation_size = 0.20
seed = 7
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=validation_size, random_state=seed)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
mod_log=LogisticRegression()
mod_kn=KNeighborsClassifier()
```

```python
mod_log.fit(x_train,y_train)
mod_kn.fit(x_train,y_train)
predictions=mod_log.predict(x_test)
predictions2=mod_kn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
accuracy=confusion_matrix(y_test,predictions)
accuracy2=confusion_matrix(y_test,predictions2)
```

```python
print(accuracy)
print(accuracy2)
```
